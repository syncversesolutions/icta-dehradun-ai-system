```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import json
import pandas as pd

from config.paths import PATHS

from domains.accommodation.signals.accommodation_signal_engine import (
    AccommodationSignalEngine
)

# ============================================
# LOAD KPI DATASET
# ============================================

input_path = (

    PATHS["accommodation"]["kpi"] +
    "accommodation_kpi.csv"
)

df = pd.read_csv(input_path)

print("\nAccommodation KPI dataset loaded ✅")

print("\nDataset Shape:")
print(df.shape)

# ============================================
# INITIALIZE SIGNAL ENGINE
# ============================================

signal_engine = (

    AccommodationSignalEngine(df)
)

# ============================================
# GENERATE SIGNALS
# ============================================

signals = signal_engine.run()

# ============================================
# SAVE SIGNALS
# ============================================

output_path = (

    PATHS["accommodation"]["signals"] +
    "signals.json"
)

with open(output_path, "w") as f:

    json.dump(
        signals,
        f,
        indent=4
    )

# ============================================
# OUTPUT
# ============================================

print("\nSignals saved at:")

print(output_path)

print("\nGenerated Signals:")

for signal in signals:

    print("\n", signal)

print("\nAccommodation signal generation complete ✅")
```
