```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import json
import pandas as pd

from datetime import datetime

from config.paths import PATHS

from domains.accommodation.modules.kpi.accommodation_kpi_engine import (
    AccommodationKPIEngine
)

# ============================================
# LOAD VALIDATED DATA
# ============================================

input_path = (

    PATHS["accommodation"]["validated"] +
    "validated_accommodation.csv"
)

df = pd.read_csv(input_path)

print("\nValidated accommodation dataset loaded ✅")

print("\nDataset Shape:")
print(df.shape)

# ============================================
# KPI ENGINE INITIALIZATION
# ============================================

kpi_engine = AccommodationKPIEngine(df)

# ============================================
# RUN KPI ENGINEERING
# ============================================

kpi_df = kpi_engine.run()

# ============================================
# KPI AGGREGATIONS
# ============================================

avg_occupancy = (
    kpi_df["occupancy_rate"].mean()
)

avg_utilization = (
    kpi_df["capacity_utilization"].mean()
)

avg_booking_velocity = (
    kpi_df["booking_velocity"].mean()
)

avg_checkin_flow = (
    kpi_df["checkin_flow"].mean()
)

# ============================================
# SAVE KPI DATASET
# ============================================

output_path = (

    PATHS["accommodation"]["kpi"] +
    "accommodation_kpi.csv"
)

kpi_df.to_csv(
    output_path,
    index=False
)

print("\nAccommodation KPI dataset saved ✅")

print(output_path)

# ============================================
# SIGNAL GENERATION
# ============================================

signals = [

    {
        "signal":
            "high_occupancy",

        "value":
            float(avg_occupancy),

        "threshold":
            0.85,

        "status":
            "critical"
            if avg_occupancy > 0.85
            else "normal",

        "timestamp":
            str(datetime.now())
    },

    {
        "signal":
            "overcapacity_risk",

        "value":
            float(avg_utilization),

        "threshold":
            0.90,

        "status":
            "critical"
            if avg_utilization > 0.90
            else "normal",

        "timestamp":
            str(datetime.now())
    }
]

# ============================================
# SAVE SIGNALS
# ============================================

signal_path = (

    PATHS["accommodation"]["signals"] +
    "signals.json"
)

with open(signal_path, "w") as f:

    json.dump(
        signals,
        f,
        indent=4
    )

print("\nAccommodation signals saved ✅")

print(signal_path)

# ============================================
# FINAL OUTPUT
# ============================================

print("\n")
print("=" * 60)

print("ACCOMMODATION KPI ENGINEERING COMPLETE")

print("=" * 60)

print("\nAverage Occupancy Rate:")
print(avg_occupancy)

print("\nAverage Capacity Utilization:")
print(avg_utilization)

print("\nAverage Booking Velocity:")
print(avg_booking_velocity)

print("\nAverage Check-in Flow:")
print(avg_checkin_flow)

print("\nFinal Dataset Sample:")
print(kpi_df.head())
```
