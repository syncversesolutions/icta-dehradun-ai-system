```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

from config.paths import PATHS

from domains.accommodation.modules.validation.accommodation_validation import (
    AccommodationValidation
)

# ============================================
# INPUT PATH
# ============================================

input_path = (

    PATHS["accommodation"]["processed"] +
    "cleaned_accommodation.csv"
)

print("\nInput Path:")
print(input_path)

# ============================================
# INITIALIZE VALIDATOR
# ============================================

validator = AccommodationValidation()

# ============================================
# LOAD DATASET
# ============================================

validator.load_dataset(input_path)

# ============================================
# RUN VALIDATION
# ============================================

validated_df = validator.run_validation()

# ============================================
# SAVE VALIDATED DATASET
# ============================================

output_path = (

    PATHS["accommodation"]["validated"] +
    "validated_accommodation.csv"
)

validated_df.to_csv(
    output_path,
    index=False
)

# ============================================
# OUTPUT
# ============================================

print("\nValidated dataset saved at:")

print(output_path)

print("\nValidated Dataset Sample:")
print(validated_df.head())

print("\nAccommodation validation complete ✅")
```
