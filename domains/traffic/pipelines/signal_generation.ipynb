```python
# ============================================
# TRAFFIC SIGNAL GENERATION PIPELINE
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import pandas as pd

from config.paths import PATHS

from domains.traffic.modules.signals.traffic_signals import (

    generate_traffic_signals,
    save_signals
)

# ============================================
# LOAD PREDICTIONS
# ============================================

input_path = (

    PATHS["traffic"]["predictions"] +
    "traffic_predictions.csv"
)

df = pd.read_csv(
    input_path
)

print("\nPrediction Dataset Shape:")
print(df.shape)

# ============================================
# GENERATE SIGNALS
# ============================================

signals = generate_traffic_signals(df)

print(
    f"\nGenerated Signals: "
    f"{len(signals)}"
)

# ============================================
# SAVE SIGNALS
# ============================================

output_path = (

    PATHS["traffic"]["signals"] +
    "signals.json"
)

save_signals(
    signals,
    output_path
)

# ============================================
# SAMPLE OUTPUT
# ============================================

print("\nSample Signals:")

for signal in signals[:5]:

    print(signal)

print("\nSignal generation complete ✅")
```
