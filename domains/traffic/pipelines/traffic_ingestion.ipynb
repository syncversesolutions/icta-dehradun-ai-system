```python
# ============================================
# TRAFFIC INGESTION PIPELINE
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

from config.paths import PATHS

from domains.traffic.modules.ingestion.traffic_ingestion import (
    ingest_traffic_data
)

# ============================================
# RAW DATA PATHS
# ============================================

checkpoint_path = (

    PATHS["traffic"]["raw"] +
    "checkpoint_logs.csv"
)

crowd_path = (

    PATHS["traffic"]["raw"] +
    "crowd_density_logs.csv"
)

print("\nCheckpoint Path:")
print(checkpoint_path)

print("\nCrowd Path:")
print(crowd_path)

# ============================================
# INGEST DATA
# ============================================

traffic_df = ingest_traffic_data(

    checkpoint_path,
    crowd_path
)

# ============================================
# SAVE CLEANED DATA
# ============================================

output_path = (

    PATHS["traffic"]["processed"] +
    "cleaned_traffic.csv"
)

traffic_df.to_csv(
    output_path,
    index=False
)

# ============================================
# OUTPUT
# ============================================

print(f"\nSaved cleaned data at:")

print(output_path)

print("\nTraffic ingestion complete ✅")
```
