```python
# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import pandas as pd
import json
import os

from datetime import datetime

from config.paths import PATHS

from domains.traffic.modules.kpi.congestion_detection import *
from domains.traffic.modules.kpi.traffic_volume import *
from domains.traffic.modules.kpi.crowd_pressure import *

# ============================================
# LOAD VALIDATED DATA
# ============================================

df = pd.read_csv(
    PATHS["traffic"]["validated"] +
    "validated_traffic.csv"
)

# ============================================
# KPI ENGINEERING
# ============================================

df = calculate_congestion_score(df)

df = calculate_traffic_volume(df)

df = calculate_crowd_pressure(df)

# ============================================
# KPI AGGREGATION
# ============================================

avg_congestion = get_avg_congestion(df)

avg_volume = get_avg_traffic_volume(df)

avg_pressure = get_avg_crowd_pressure(df)

# ============================================
# SAVE KPI DATASET
# ============================================

kpi_dir = PATHS["traffic"]["kpi"]
os.makedirs(kpi_dir, exist_ok=True)

df.to_csv(
    PATHS["traffic"]["kpi"] +
    "traffic_kpi.csv",
    index=False
)

# ============================================
# GENERATE SIGNALS
# ============================================

signals = []

signals.append({

    "kpi": "avg_congestion",

    "value": float(avg_congestion),

    "timestamp": str(datetime.now())
})

signals.append({

    "kpi": "avg_traffic_volume",

    "value": float(avg_volume),

    "timestamp": str(datetime.now())
})

signals.append({

    "kpi": "avg_crowd_pressure",

    "value": float(avg_pressure),

    "timestamp": str(datetime.now())
})

# ============================================
# SAVE SIGNALS
# ============================================

with open(
    PATHS["traffic"]["signals"] +
    "signals.json",
    "w"
) as f:

    json.dump(
        signals,
        f,
        indent=4
    )

# ============================================
# OUTPUT
# ============================================

print(df.head())
print(df)

print("Traffic KPI Engineering complete ✅")
```
