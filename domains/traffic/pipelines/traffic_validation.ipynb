```python
# ============================================
# TRAFFIC VALIDATION PIPELINE
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import pandas as pd

from config.paths import PATHS

from domains.traffic.modules.validation.traffic_validation import (
    validate_traffic_data
)

# ============================================
# LOAD CLEANED DATA
# ============================================

input_path = (

    PATHS["traffic"]["processed"] +
    "cleaned_traffic.csv"
)

print("\nLoading cleaned traffic dataset...")

print(input_path)

df = pd.read_csv(
    input_path
)

# ============================================
# DEBUG INPUT
# ============================================

print("\nInput Dataset Shape:")
print(df.shape)

print("\nInput Dataset Columns:")
print(df.columns.tolist())

print("\nInput Dataset Sample:")
print(df.head())

# ============================================
# VALIDATE DATA
# ============================================

validated_df = validate_traffic_data(df)

# ============================================
# SAVE VALIDATED DATA
# ============================================

output_path = (

    PATHS["traffic"]["validated"] +
    "validated_traffic.csv"
)

validated_df.to_csv(
    output_path,
    index=False
)

# ============================================
# OUTPUT
# ============================================

print("\nValidated Dataset Shape:")
print(validated_df.shape)

print(f"\nValidated data saved at:")

print(output_path)

print("\nTraffic validation complete ✅")
```
