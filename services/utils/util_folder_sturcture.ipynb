```python
import os

project_root = '/content/drive/MyDrive/project_cd'

print(f"Folder structure for: {project_root}")
for root, dirs, files in os.walk(project_root):
    # Exclude .git directory from traversal
    if '.git' in dirs:
        dirs.remove('.git')

    level = root.replace(project_root, '').count(os.sep)
    indent = ' ' * 4 * (level)

    # Only print directories that are not excluded
    if os.path.basename(root) != '.git':
        print(f'{indent}{os.path.basename(root)}/')

    subindent = ' ' * 4 * (level + 1)
    excluded_files = ['data.zip', 'project_cd-20260524T105613Z-3-001.zip']
    for f in files:
        if f not in excluded_files:
            print(f'{subindent}{f}')
```
