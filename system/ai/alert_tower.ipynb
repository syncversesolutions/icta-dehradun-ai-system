```python
# ============================================
# AI ALERT TOWER
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import json

from config.paths import PATHS

from system.ai.alert_engine import (
    generate_alerts
)

# ============================================
# LOAD IMPACT ANALYSIS
# ============================================

impact_path = (

    PATHS["system"]["graph"] +
    "impact_analysis.json"
)

with open(
    impact_path,
    "r"
) as file:

    impact_analysis = json.load(file)

print(
    f"\nImpact Analyses Loaded: "
    f"{len(impact_analysis)}"
)

# ============================================
# GENERATE ALERTS
# ============================================

alerts = generate_alerts(
    impact_analysis
)

# ============================================
# SORT ALERTS
# ============================================

alerts = sorted(

    alerts,

    key=lambda x:
        x["severity_score"],

    reverse=True
)

# ============================================
# SAVE ACTIVE ALERTS
# ============================================

output_path = (

    PATHS["system"]["ai"] +
    "active_alerts.json"
)

with open(
    output_path,
    "w"
) as file:

    json.dump(
        alerts,
        file,
        indent=4
    )

# ============================================
# OUTPUT
# ============================================

print(
    f"\nTotal Active Alerts: "
    f"{len(alerts)}"
)

print("\nTop Alerts:\n")

for alert in alerts[:5]:

    print(alert)

print(f"\nActive alerts saved at:")

print(output_path)

print("\nAI Alert Tower complete ✅")
```
