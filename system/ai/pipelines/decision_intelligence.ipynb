```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import pandas as pd

from config.paths import PATHS

from system.state.state_manager import (
    StateManager
)

from system.prediction.risk_forecaster import (
    RiskForecaster
)

from system.prediction.trend_analyzer import (
    TrendAnalyzer
)

from system.ai.decision.decision_engine import (
    DecisionEngine
)

# ============================================
# LOAD STATE
# ============================================

state_manager = StateManager()

state = state_manager.load_state()

# ============================================
# LOAD KPI DATA
# ============================================

kpi_path = (

    PATHS["accommodation"]["kpi"] +
    "accommodation_kpi.csv"
)

df = pd.read_csv(kpi_path)

# ============================================
# GENERATE PREDICTIONS
# ============================================

forecaster = RiskForecaster(
    state
)

predictions = forecaster.run()

# ============================================
# ANALYZE TRENDS
# ============================================

trend_analyzer = TrendAnalyzer(df)

trends = trend_analyzer.analyze()

# ============================================
# DECISION INTELLIGENCE
# ============================================

decision_engine = DecisionEngine()

decision_output = (

    decision_engine.run(

        state,
        predictions,
        trends
    )
)

# ============================================
# OUTPUT
# ============================================

print("\n")
print("=" * 60)

print("ICTA DECISION INTELLIGENCE")

print("=" * 60)

print("\nPolicy Actions:")

for action in decision_output[
    "policy_actions"
]:

    print(f"   ✓ {action}")

print("\nRecommendations:")

for item in decision_output[
    "recommendations"
]:

    print("\n", item)

print("\nReasoning:")
print(
    decision_output[
        "reasoning"
    ]
)

print("\n")
print("=" * 60)
```
