```python
# ============================================
# AI IMPACT ANALYSIS ENGINE
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import json

from config.paths import PATHS

from system.graph.graph_utils import (

    load_graph,
    analyze_signal_impacts
)

# ============================================
# LOAD SIGNALS
# ============================================

signal_path = (

    PATHS["system"]["signals"] +
    "all_signals.json"
)

with open(
    signal_path,
    "r"
) as file:

    all_signals = json.load(file)

print(
    f"\nSignals Loaded: "
    f"{len(all_signals)}"
)

# ============================================
# LOAD DEPENDENCY GRAPH
# ============================================

graph_path = (

    PATHS["system"]["graph"] +
    "dependency_graph.json"
)

graph = load_graph(
    graph_path
)

print("\nDependency graph loaded ✅")

# ============================================
# ANALYZE IMPACTS
# ============================================

impact_analysis = []

for signal in all_signals:

    result = analyze_signal_impacts(

        graph,
        signal
    )

    impact_analysis.append(
        result
    )

# ============================================
# SAVE IMPACT ANALYSIS
# ============================================

output_path = (

    PATHS["system"]["graph"] +
    "impact_analysis.json"
)

with open(
    output_path,
    "w"
) as file:

    json.dump(
        impact_analysis,
        file,
        indent=4
    )

# ============================================
# OUTPUT
# ============================================

print(
    f"\nImpact Analyses Generated: "
    f"{len(impact_analysis)}"
)

print("\nSample Impact Analysis:\n")

for item in impact_analysis[:3]:

    print(item)

print(f"\nImpact analysis saved at:")

print(output_path)

print("\nAI impact analysis complete ✅")
```
