```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

from system.knowledge.modules.graph.graph_engine import (
    GraphEngine
)

from system.knowledge.modules.ontology.ontology_engine import (
    OntologyEngine
)

from system.knowledge.modules.relationships.relationship_manager import (
    RelationshipManager
)

from system.knowledge.modules.reasoning.knowledge_reasoner import (
    KnowledgeReasoner
)

# ============================================
# INITIALIZE COMPONENTS
# ============================================

graph_engine = GraphEngine()

ontology_engine = OntologyEngine()

relationship_manager = (
    RelationshipManager()
)

reasoner = KnowledgeReasoner()

# ============================================
# REGISTER RELATIONSHIPS
# ============================================

relationship_manager.register(

    "accommodation",

    "traffic",

    "increases_congestion"
)

relationship_manager.register(

    "crowd",

    "health",

    "increases_medical_pressure"
)

# ============================================
# BUILD KNOWLEDGE GRAPH
# ============================================

graph_engine.add_relationship(

    "accommodation",

    "traffic",

    "increases_congestion"
)

graph_engine.add_relationship(

    "crowd",

    "health",

    "increases_medical_pressure"
)

# ============================================
# CONNECTED NODES
# ============================================

connected = (

    graph_engine.connected_nodes(
        "accommodation"
    )
)

# ============================================
# ONTOLOGY LOOKUP
# ============================================

definition = (

    ontology_engine.get_definition(
        "traffic"
    )
)

# ============================================
# KNOWLEDGE REASONING
# ============================================

reasoning = (

    reasoner.reason(

        signal=
            "accommodation.occupancy_rate",

        connected_nodes=
            connected
    )
)

# ============================================
# OUTPUT
# ============================================

graph_engine.display_graph()

print("\nTraffic Ontology:")
print(definition)

print("\nReasoning:")
print(reasoning)
```
