```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import pandas as pd

from config.paths import PATHS

from system.state.state_manager import (
    StateManager
)

from system.prediction.modules.forecasting.risk_forecaster import (
    RiskForecaster
)

from system.prediction.modules.trends.trend_analyzer import (
    TrendAnalyzer
)

from system.prediction.modules.simulation.simulation_engine import (
    SimulationEngine
)

# ============================================
# LOAD SYSTEM STATE
# ============================================

state_manager = StateManager()

state = state_manager.load_state()

print("\nSystem state loaded ✅")

# ============================================
# LOAD KPI DATA
# ============================================

kpi_path = (

    PATHS["accommodation"]["kpi"] +
    "accommodation_kpi.csv"
)

df = pd.read_csv(kpi_path)

print("\nKPI dataset loaded ✅")

# ============================================
# RISK FORECASTING
# ============================================

forecaster = RiskForecaster(
    state
)

predictions = forecaster.run()

# ============================================
# TREND ANALYSIS
# ============================================

trend_analyzer = TrendAnalyzer(df)

trends = trend_analyzer.analyze()

# ============================================
# SCENARIO SIMULATION
# ============================================

simulator = SimulationEngine()

scenarios = simulator.run()

# ============================================
# OUTPUT
# ============================================

print("\n")
print("=" * 60)

print("ICTA PREDICTION SYSTEM")

print("=" * 60)

print("\nPredictions:")

for item in predictions:

    print("\n", item)

print("\nTrend Analysis:")
print(trends)

print("\nSimulated Scenarios:")

for scenario in scenarios:

    print("\n", scenario)

print("\n")
print("=" * 60)
```
