```python
# ============================================
# ENVIRONMENT INITIALIZATION
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

from system.realtime.modules.listener.event_listener import (
    EventListener
)

from system.realtime.modules.bus.live_signal_bus import (
    LiveSignalBus
)

from system.realtime.modules.processor.stream_processor import (
    StreamProcessor
)

from system.realtime.modules.monitor.realtime_monitor import (
    RealtimeMonitor
)

# ============================================
# INITIALIZE COMPONENTS
# ============================================

listener = EventListener()

signal_bus = LiveSignalBus()

processor = StreamProcessor()

monitor = RealtimeMonitor()

# ============================================
# RECEIVE REALTIME EVENT
# ============================================

event = listener.receive_event(

    domain="accommodation",

    event_type="occupancy_update",

    payload={

        "occupancy_rate":
            0.92,

        "status":
            "critical"
    }
)

# ============================================
# BUILD LIVE SIGNAL
# ============================================

signal = {

    "domain":
        "accommodation",

    "signal":
        "high_occupancy",

    "status":
        "critical",

    "value":
        0.92
}

# ============================================
# PUBLISH SIGNAL
# ============================================

signal_bus.publish_signal(
    signal
)

# ============================================
# PROCESS LIVE STREAM
# ============================================

stream = signal_bus.get_stream()

processor.process_stream(
    stream
)

# ============================================
# UPDATE MONITOR
# ============================================

monitor.update_metrics(

    events=1,

    critical=1,

    streams=len(stream)
)

# ============================================
# DISPLAY RESULTS
# ============================================

listener.display_events()

signal_bus.display_stream()

monitor.display_metrics()
```
