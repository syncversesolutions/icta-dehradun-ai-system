```python
# ============================================
# CENTRAL SIGNAL AGGREGATION
# ============================================

# ============================================
# ENVIRONMENT
# ============================================

%run /content/drive/MyDrive/project_cd/config/environment.py

# ============================================
# IMPORTS
# ============================================

import json

from config.paths import PATHS

# ============================================
# LOAD TRAFFIC SIGNALS
# ============================================

traffic_signal_path = (

    PATHS["traffic"]["signals"] +
    "signals.json"
)

with open(
    traffic_signal_path,
    "r"
) as file:

    traffic_signals = json.load(file)

print(
    f"\nTraffic Signals Loaded: "
    f"{len(traffic_signals)}"
)

# ============================================
# INITIALIZE CENTRAL SIGNAL BUS
# ============================================

all_signals = []

# ============================================
# AGGREGATE SIGNALS
# ============================================

all_signals.extend(
    traffic_signals
)

# ============================================
# SAVE CENTRAL SIGNAL BUS
# ============================================

output_path = (

    PATHS["system"]["signals"] +
    "all_signals.json"
)

with open(
    output_path,
    "w"
) as file:

    json.dump(
        all_signals,
        file,
        indent=4
    )

# ============================================
# OUTPUT
# ============================================

print(
    f"\nTotal Aggregated Signals: "
    f"{len(all_signals)}"
)

print("\nSample Signals:")

for signal in all_signals[:5]:

    print(signal)

print(f"\nAll signals saved at:")

print(output_path)

print("\nCentral signal aggregation complete ✅")
```
